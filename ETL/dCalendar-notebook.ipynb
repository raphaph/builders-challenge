```python
import pandas as pd
import datetime
import calendar

data_inicial = datetime.date(2020, 1, 1)
data_final = datetime.date(2022, 12, 31)
dias_calendario = pd.date_range(start=data_inicial, end=data_final, freq='D')

calendario_df = pd.DataFrame(dias_calendario, columns=['Data'])

calendario_df['dia'] = calendario_df['Data'].dt.day
calendario_df['mes'] = calendario_df['Data'].dt.month
calendario_df['mes_nome'] = calendario_df['Data'].dt.strftime('%b')
calendario_df['ano'] = calendario_df['Data'].dt.year
calendario_df['semana_do_ano'] = calendario_df['Data'].dt.isocalendar().week


calendario_df['ano_mes'] = calendario_df['Data'].dt.strftime('%Y%m')
calendario_df['ano_semana'] = calendario_df['Data'].dt.strftime('%Y') + calendario_df['semana_do_ano'].astype(str).str.zfill(2)
calendario_df['mes_ano'] = calendario_df['Data'].dt.strftime('%b/%Y')

calendario_df['mes_ano'] = calendario_df['mes_ano'].str.lower().replace(calendar.month_abbr[0].lower(), calendar.month_abbr[12].lower())
calendario_df.to_csv('dCalendar.csv',index=False)


display(calendario_df)
```
